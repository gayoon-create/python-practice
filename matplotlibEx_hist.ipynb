```python
hist : 히스토그램 -> 자료의 분포 상태를 직사각형 모양의 막대 그래프로 나타낸 것
```

```python
import matplotlib.pyplot as plt

# x축 자료의 빈도수가 y축으로 자동 결정된다.
plt.hist([1, 1, 2, 3, 4, 4, 5, 6, 7, 7, 7, 8, 9, 10])
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

dice = []
for i in range(5) :
    dice.append(random.randint(1, 6))

plt.hist(dice)
plt.show()
```
