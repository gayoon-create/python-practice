pie 차트(원 그래프) : 전체 데이터 중 특정 데이터의 비율(퍼센테이지)을 구해 시각화한 것

```python
import matplotlib.pyplot as plt

plt.rc('font', family='D2Coding') 
plt.title('우리반 혈액형 분포도')
label = ['A형', 'B형', 'O형', 'AB형']  # 각 혈액형의 이름
colors = ['red', 'blue', 'green', 'yellow'] 
plt.pie([3, 1, 4, 1], labels = label, autopct='%.1f%%', explode=(0, 1, 0, 0), colors = colors)
# explode : 돌출
# autopct : 각 차트의 퍼센테이지값 표시

plt.legend()
plt.show()
```
